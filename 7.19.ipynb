## while 循环
- 当一个条件保持真的时候while循环重复执行语句
- while 循环一定要有结束条件，否则很容易进入死循环
- while 循环的语法是：

    while loop-contunuation-conndition:

        Statement

```python
i = 0
while i<10:
    i +=1
    print('')
```

```python
i = 0
a = 'abcd'
while i <4:
    
    print(a[i])
    i+=1
     
```

## 嵌套循环
- 一个循环可以嵌套另一个循环
- 每次循环外层时，内层循环都会被刷新重新完成循环
- 也就是说，大循环执行一次，小循环会全部执行一次
- 注意：
> - 多层循环非常耗时
  - 最多使用3层循环

## EP：
- 使用多层循环完成9X9乘法表
- 显示50以内所有的素数

```python
s=[]
i=1
for i in range(1,10):
    j=1
    for j in range(1,i+1):
        print('%d*%d=%d '%(i,j,i*j),end='')
    print()
```

```python
for i in range(2,50):
    for j in range(2,i):
        if i%j==0 :
            break
    else:
        print(i,end=' ')
```
